```python
movie_freq={}
```

```python
for i in range(len(ratings)):
    if ratings['movieId'][i] in movie_freq:
        movie_freq[ratings['movieId'][i]]+=1
    else:
        movie_freq[ratings['movieId'][i]]=1
```
